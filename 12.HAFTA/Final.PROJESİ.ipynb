```python
class TarimUrunu:
    def __init__(self):
        self.urunler = ["1 = Karpuz", "2 = Çilek", "3 = Kavun", "4 = Kayısı", "5 = Kivi", "6 = Patates", "7 = Soğan", "8 = Fasulye", "9 = Havuç", "10 = Ispanak"]
        self.ph = None
        self.nem = None
        self.toprak_sicaklik = None
        self.hava_sicaklik = None

    def urun_sec(self):
        while True:
            urun_no = input("Lütfen istediğiniz ürünün numarasını giriniz: ")
            if urun_no == "q":
                print("Çıkış yapılıyor...")
                return None
            elif urun_no.isdigit() and int(urun_no) in range(1, 11):
                secilen_urun = self.urunler[int(urun_no) - 1].split(" = ")[1]
                print(f"Seçtiğiniz ürün: {secilen_urun}")
                break
            else:
                print("Hatalı giriş yaptınız. Lütfen 1-10 arasında bir sayı giriniz veya q tuşuna basarak çıkış yapınız.")



    def toprak_verilerini_gir(self):
        self.ph = float(input("Toprak pH değerini girin: "))
        self.nem = float(input("Toprak nem değerini girin: "))
        self.toprak_sicaklik = float(input("Toprak sıcaklık değerini girin: "))

    def hava_verilerini_gir(self):
        self.hava_sicaklik = float(input("Hava sıcaklık değerini girin: "))
    

def main():
    tarim_urunu = TarimUrunu()
    urun_no = tarim_urunu.urun_sec()

    if urun_no == 1:
        karpuz_yetistirme = KarpuzYetistirme()
        karpuz_yetistirme.bilgi_gir(tarim_urunu.ph, tarim_urunu.nem, tarim_urunu.toprak_sicaklik, tarim_urunu.hava_sicaklik)
        karpuz_yetistirme.yetistir()
    elif urun_no == 2:
        cilek_yetistirme = CilekYetistirme()
        cilek_yetistirme.bilgi_gir(tarim_urunu.ph, tarim_urunu.nem, tarim_urunu.toprak_sicaklik, tarim_urunu.hava_sicaklik)
        cilek_yetistirme.yetistir()
    elif urun_no == 3:
        kavun_yetistirme = KavunYetistirme()
        kavun_yetistirme.bilgi_gir(tarim_urunu.ph, tarim_urunu.nem, tarim_urunu.toprak_sicaklik, tarim_urunu.hava_sicaklik)
        kavun_yetistirme.yetistir()
    elif urun_no == 4:
        kayisi_yetistirme = KayisiYetistirme()
        kayisi_yetistirme.bilgi_gir(tarim_urunu.ph, tarim_urunu.nem, tarim_urunu.toprak_sicaklik, tarim_urunu.hava_sicaklik)
        kayisi_yetistirme.yetistir()
    elif urun_no == 5:
        kivi_yetistirme = KiviYetistirme()
        kivi_yetistirme.bilgi_gir(tarim_urunu.ph, tarim_urunu.nem, tarim_urunu.toprak_sicaklik, tarim_urunu.hava_sicaklik)
        kivi_yetistirme.yetistir()
    elif urun_no == 6:
        patates_yetistirme = PatatesYetistirme()
        patates_yetistirme.bilgi_gir(tarim_urunu.ph, tarim_urunu.nem, tarim_urunu.toprak_sicaklik, tarim_urunu.hava_sicaklik)
        patates_yetistirme.yetistir()
    elif urun_no == 7:
        sogan_yetistirme = SoganYetistirme()
        sogan_yetistirme.bilgi_gir(tarim_urunu.ph, tarim_urunu.nem, tarim_urunu.toprak_sicaklik, tarim_urunu.hava_sicaklik)
        sogan_yetistirme.yetistir()
    elif urun_no == 8:
        fasulye_yetistirme = FasulyeYetistirme()
        fasulye_yetistirme.bilgi_gir(tarim_urunu.ph, tarim_urunu.nem, tarim_urunu.toprak_sicaklik, tarim_urunu.hava_sicaklik)
        fasulye_yetistirme.yetistir()     


class KarpuzYetistirme(TarimUrunu):
    def __init__(self, toprak_pH, toprak_nem, toprak_sicaklik, hava_sicaklik):
        self.toprak_pH = toprak_pH
        self.toprak_nem = toprak_nem
        self.toprak_sicaklik = toprak_sicaklik
        self.hava_sicaklik = hava_sicaklik
        
    def toprak_pH_kontrol(self):
        if self.toprak_pH >= 6.0 and self.toprak_pH <= 6.8:
            print("Toprak pH değeri uygun.")
        elif self.toprak_pH > 6.8:
            print("Toprak pH değeri yüksek. Toprağa kireç veya kül ekleyebilirsiniz.")
        elif self.toprak_pH < 6.0:
            print("Toprak pH değeri düşük. Sulama suyuna asitlik veren malzemeler veya organik malzemeler ekleyebilirsiniz.")
    
    def toprak_nem_kontrol(self):
        if self.toprak_nem >= 60 and self.toprak_nem <= 80:
            print("Toprak nem değeri uygun.")
        elif self.toprak_nem < 60:
            print("Toprak nem değeri düşük. Daha sık sulama yapabilirsiniz.")
        elif self.toprak_nem > 80:
            print("Toprak nem değeri yüksek. Daha az sulama yapabilirsiniz.")
    
    def toprak_sicaklik_kontrol(self):
        if self.toprak_sicaklik >= 18 and self.toprak_sicaklik <= 32:
            print("Toprak sıcaklık değeri uygun.")
        elif self.toprak_sicaklik < 18:
            print("Toprak sıcaklık değeri düşük. Isıtıcı kullanabilirsiniz.")
        elif self.toprak_sicaklik > 32:
            print("Toprak sıcaklık değeri yüksek. Gölgeleme yapabilirsiniz.")
    
    def hava_sicaklik_kontrol(self):
        if self.hava_sicaklik >= 25 and self.hava_sicaklik <= 35:
            print("Hava sıcaklık değeri uygun.")
        elif self.hava_sicaklik < 25:
            print("Hava sıcaklık değeri düşük. Seranın havalandırmasını açabilirsiniz.")
        elif self.hava_sicaklik > 35:
            print("Hava sıcaklık değeri yüksek. Seranın havalandırmasını artırabilirsiniz.")
        
    def karpuz_yetistirme_kontrol(self):
        self.toprak_pH_kontrol()
        self.toprak_nem_kontrol()
        self.toprak_sicaklik_kontrol()
        self.hava_sicaklik_kontrol()
      
       
class CilekYetistirme(TarimUrunu):
    def __init__(self, toprak_pH, toprak_nem, toprak_sicaklik, hava_sicaklik):
        self.toprak_pH = toprak_pH
        self.toprak_nem = toprak_nem
        self.toprak_sicaklik = toprak_sicaklik
        self.hava_sicaklik = hava_sicaklik
        
    def toprak_pH_kontrol(self):
        if self.toprak_pH >= 5.5 and self.toprak_pH <= 6.5:
            print("Toprak pH değeri uygun.")
        elif self.toprak_pH > 6.5:
            print("Toprak pH değeri yüksek. Toprağa turba veya sülfürik asit ekleyebilirsiniz.")
        elif self.toprak_pH < 5.5:
            print("Toprak pH değeri düşük. Toprağa kireç veya dolomit ekleyebilirsiniz.")
    
    def toprak_nem_kontrol(self):
        if self.toprak_nem >= 70 and self.toprak_nem <= 80:
            print("Toprak nem değeri uygun.")
        elif self.toprak_nem < 70:
            print("Toprak nem değeri düşük. Daha sık sulama yapabilirsiniz.")
        elif self.toprak_nem > 80:
            print("Toprak nem değeri yüksek. Daha az sulama yapabilirsiniz.")
    
    def toprak_sicaklik_kontrol(self):
        if self.toprak_sicaklik >= 15 and self.toprak_sicaklik <= 25:
            print("Toprak sıcaklık değeri uygun.")
        elif self.toprak_sicaklik < 15:
            print("Toprak sıcaklık değeri düşük. Isıtıcı kullanabilirsiniz.")
        elif self.toprak_sicaklik > 25:
            print("Toprak sıcaklık değeri yüksek. Gölgeleme yapabilirsiniz.")
    
    def hava_sicaklik_kontrol(self):
        if self.hava_sicaklik >= 15 and self.hava_sicaklik <= 25:
            print("Hava sıcaklık değeri uygun.")
        elif self.hava_sicaklik < 15:
            print("Hava sıcaklık değeri düşük. Seranın havalandırmasını açabilirsiniz.")
        elif self.hava_sicaklik > 25:
            print("Hava sıcaklık değeri yüksek. Seranın havalandırmasını artırabilirsiniz.")
        
    def cilek_yetistirme_kontrol(self):
        self.toprak_pH_kontrol()
        self.toprak_nem_kontrol()
        self.toprak_sicaklik_kontrol()
        self.hava_sicaklik_kontrol()




class KavunYetistirme:
    def __init__(self, toprak_pH, toprak_nem, toprak_sicaklik, hava_sicaklik):
        self.toprak_pH = toprak_pH
        self.toprak_nem = toprak_nem
        self.toprak_sicaklik = toprak_sicaklik
        self.hava_sicaklik = hava_sicaklik
        
    def toprak_pH_kontrol(self):
        if self.toprak_pH >= 6.0 and self.toprak_pH <= 6.8:
            print("Toprak pH değeri uygun.")
        elif self.toprak_pH > 6.8:
            print("Toprak pH değeri yüksek. Toprağa kireç veya kül ekleyebilirsiniz.")
        elif self.toprak_pH < 6.0:
            print("Toprak pH değeri düşük. Sulama suyuna asitlik veren malzemeler veya organik malzemeler ekleyebilirsiniz.")
    
    def toprak_nem_kontrol(self):
        if self.toprak_nem >= 60 and self.toprak_nem <= 80:
            print("Toprak nem değeri uygun.")
        elif self.toprak_nem < 60:
            print("Toprak nem değeri düşük. Daha sık sulama yapabilirsiniz.")
        elif self.toprak_nem > 80:
            print("Toprak nem değeri yüksek. Daha az sulama yapabilirsiniz.")
    
    def toprak_sicaklik_kontrol(self):
        if self.toprak_sicaklik >= 22 and self.toprak_sicaklik <= 30:
            print("Toprak sıcaklık değeri uygun.")
        elif self.toprak_sicaklik < 22:
            print("Toprak sıcaklık değeri düşük. Isıtıcı kullanabilirsiniz.")
        elif self.toprak_sicaklik > 30:
            print("Toprak sıcaklık değeri yüksek. Gölgeleme yapabilirsiniz.")
    
    def hava_sicaklik_kontrol(self):
        if self.hava_sicaklik >= 25 and self.hava_sicaklik <= 35:
            print("Hava sıcaklık değeri uygun.")
        elif self.hava_sicaklik < 25:
            print("Hava sıcaklık değeri düşük. Seranın havalandırmasını açabilirsiniz.")
        elif self.hava_sicaklik > 35:
            print("Hava sıcaklık değeri yüksek. Seranın havalandırmasını artırabilirsiniz.")
        
    def kavun_yetistirme_kontrol(self):
        self.toprak_pH_kontrol()
        self.toprak_nem_kontrol()
        self.toprak_sicaklik_kontrol()
        self.hava_sicaklik_kontrol()

tarim_urunu = TarimUrunu()
tarim_urunu.urun_sec()
tarim_urunu.toprak_verilerini_gir()
tarim_urunu.hava_verilerini_gir()

kontrol = KavunYetistirme(tarim_urunu.ph,tarim_urunu.nem,tarim_urunu.toprak_sicaklik,tarim_urunu.hava_sicaklik)   
kontrol.kavun_yetistirme_kontrol() 

class KayisiYetistirme:
    def __init__(self, toprak_pH, toprak_nem, toprak_sicaklik, hava_sicaklik):
        self.toprak_pH = toprak_pH
        self.toprak_nem = toprak_nem
        self.toprak_sicaklik = toprak_sicaklik
        self.hava_sicaklik = hava_sicaklik
        
    def toprak_pH_kontrol(self):
        if self.toprak_pH >= 6.0 and self.toprak_pH <= 7.0:
            print("Toprak pH değeri uygun.")
        elif self.toprak_pH > 7.0:
            print("Toprak pH değeri yüksek. Toprağa kireç veya kül ekleyebilirsiniz.")
        elif self.toprak_pH < 6.0:
            print("Toprak pH değeri düşük. Sulama suyuna asitlik veren malzemeler veya organik malzemeler ekleyebilirsiniz.")
    
    def toprak_nem_kontrol(self):
        if self.toprak_nem >= 50 and self.toprak_nem <= 70:
            print("Toprak nem değeri uygun.")
        elif self.toprak_nem < 50:
            print("Toprak nem değeri düşük. Daha sık sulama yapabilirsiniz.")
        elif self.toprak_nem > 70:
            print("Toprak nem değeri yüksek. Daha az sulama yapabilirsiniz.")
    
    def toprak_sicaklik_kontrol(self):
        if self.toprak_sicaklik >= 20 and self.toprak_sicaklik <= 30:
            print("Toprak sıcaklık değeri uygun.")
        elif self.toprak_sicaklik < 20:
            print("Toprak sıcaklık değeri düşük. Isıtıcı kullanabilirsiniz.")
        elif self.toprak_sicaklik > 30:
            print("Toprak sıcaklık değeri yüksek. Gölgeleme yapabilirsiniz.")
    
    def hava_sicaklik_kontrol(self):
        if self.hava_sicaklik >= 20 and self.hava_sicaklik <= 35:
            print("Hava sıcaklık değeri uygun.")
        elif self.hava_sicaklik < 20:
            print("Hava sıcaklık değeri düşük. Seranın havalandırmasını açabilirsiniz.")
        elif self.hava_sicaklik > 35:
            print("Hava sıcaklık değeri yüksek. Seranın havalandırmasını artırabilirsiniz.")
        
    def kayisi_yetistirme_kontrol(self):
        self.toprak_pH_kontrol()
        self.toprak_nem_kontrol()
        self.toprak_sicaklik_kontrol()
        self.hava_sicaklik_kontrol()

tarim_urunu = TarimUrunu()
tarim_urunu.urun_sec()
tarim_urunu.toprak_verilerini_gir()
tarim_urunu.hava_verilerini_gir()

kontrol = KayisiYetistirme(tarim_urunu.ph,tarim_urunu.nem,tarim_urunu.toprak_sicaklik,tarim_urunu.hava_sicaklik)   
kontrol.kayisi_yetistirme_kontrol() 


class KiviYetistirme:
    def __init__(self, toprak_pH, toprak_nem, toprak_sicaklik, hava_sicaklik):
        self.toprak_pH = toprak_pH
        self.toprak_nem = toprak_nem
        self.toprak_sicaklik = toprak_sicaklik
        self.hava_sicaklik = hava_sicaklik
        
    def toprak_pH_kontrol(self):
        if self.toprak_pH >= 6.0 and self.toprak_pH <= 6.8:
            print("Toprak pH değeri uygun.")
        elif self.toprak_pH > 6.8:
            print("Toprak pH değeri yüksek. Toprağa kireç veya kül ekleyebilirsiniz.")
        elif self.toprak_pH < 6.0:
            print("Toprak pH değeri düşük. Sulama suyuna asitlik veren malzemeler veya organik malzemeler ekleyebilirsiniz.")
    
    def toprak_nem_kontrol(self):
        if self.toprak_nem >= 60 and self.toprak_nem <= 80:
            print("Toprak nem değeri uygun.")
        elif self.toprak_nem < 60:
            print("Toprak nem değeri düşük. Daha sık sulama yapabilirsiniz.")
        elif self.toprak_nem > 80:
            print("Toprak nem değeri yüksek. Daha az sulama yapabilirsiniz.")
    
    def toprak_sicaklik_kontrol(self):
        if self.toprak_sicaklik >= 18 and self.toprak_sicaklik <= 32:
            print("Toprak sıcaklık değeri uygun.")
        elif self.toprak_sicaklik < 18:
            print("Toprak sıcaklık değeri düşük. Isıtıcı kullanabilirsiniz.")
        elif self.toprak_sicaklik > 32:
            print("Toprak sıcaklık değeri yüksek. Gölgeleme yapabilirsiniz.")
    
    def hava_sicaklik_kontrol(self):
        if self.hava_sicaklik >= 25 and self.hava_sicaklik <= 35:
            print("Hava sıcaklık değeri uygun.")
        elif self.hava_sicaklik < 25:
            print("Hava sıcaklık değeri düşük. Seranın havalandırmasını açabilirsiniz.")
        elif self.hava_sicaklik > 35:
            print("Hava sıcaklık değeri yüksek. Seranın havalandırmasını artırabilirsiniz.")
        
    def kivi_yetistirme_kontrol(self):
        self.toprak_pH_kontrol()
        self.toprak_nem_kontrol()
        self.toprak_sicaklik_kontrol()
        self.hava_sicaklik_kontrol()

tarim_urunu = TarimUrunu()
tarim_urunu.urun_sec()
tarim_urunu.toprak_verilerini_gir()
tarim_urunu.hava_verilerini_gir()

kontrol = KiviYetistirme(tarim_urunu.ph,tarim_urunu.nem,tarim_urunu.toprak_sicaklik,tarim_urunu.hava_sicaklik)   
kontrol.kivi_yetistirme_kontrol() 
 

class PatatesYetistirme:
    def __init__(self, toprak_pH, toprak_nem, toprak_sicaklik, hava_sicaklik):
        self.toprak_pH = toprak_pH
        self.toprak_nem = toprak_nem
        self.toprak_sicaklik = toprak_sicaklik
        self.hava_sicaklik = hava_sicaklik
        
    def toprak_pH_kontrol(self):
        if self.toprak_pH >= 5.5 and self.toprak_pH <= 7.0:
            print("Toprak pH değeri uygun.")
        elif self.toprak_pH > 7.0:
            print("Toprak pH değeri yüksek. Toprağa kireç veya kül ekleyebilirsiniz.")
        elif self.toprak_pH < 5.5:
            print("Toprak pH değeri düşük. Sulama suyuna asitlik veren malzemeler veya organik malzemeler ekleyebilirsiniz.")
    
    def toprak_nem_kontrol(self):
        if self.toprak_nem >= 70 and self.toprak_nem <= 80:
            print("Toprak nem değeri uygun.")
        elif self.toprak_nem < 70:
            print("Toprak nem değeri düşük. Daha sık sulama yapabilirsiniz.")
        elif self.toprak_nem > 80:
            print("Toprak nem değeri yüksek. Daha az sulama yapabilirsiniz.")
    
    def toprak_sicaklik_kontrol(self):
        if self.toprak_sicaklik >= 15 and self.toprak_sicaklik <= 25:
            print("Toprak sıcaklık değeri uygun.")
        elif self.toprak_sicaklik < 15:
            print("Toprak sıcaklık değeri düşük. Isıtıcı kullanabilirsiniz.")
        elif self.toprak_sicaklik > 25:
            print("Toprak sıcaklık değeri yüksek. Gölgeleme yapabilirsiniz.")
    
    def hava_sicaklik_kontrol(self):
        if self.hava_sicaklik >= 20 and self.hava_sicaklik <= 30:
            print("Hava sıcaklık değeri uygun.")
        elif self.hava_sicaklik < 20:
            print("Hava sıcaklık değeri düşük. Seranın havalandırmasını açabilirsiniz.")
        elif self.hava_sicaklik > 30:
            print("Hava sıcaklık değeri yüksek. Seranın havalandırmasını artırabilirsiniz.")
        
    def patates_yetistirme_kontrol(self):
        self.toprak_pH_kontrol()
        self.toprak_nem_kontrol()
        self.toprak_sicaklik_kontrol()
        self.hava_sicaklik_kontrol()



class SoganYetistirme:
    def __init__(self, toprak_pH, toprak_nem, toprak_sicaklik, hava_sicaklik):
        self.toprak_pH = toprak_pH
        self.toprak_nem = toprak_nem
        self.toprak_sicaklik = toprak_sicaklik
        self.hava_sicaklik = hava_sicaklik
        
    def toprak_pH_kontrol(self):
        if self.toprak_pH >= 6.0 and self.toprak_pH <= 7.0:
            print("Toprak pH değeri uygun.")
        elif self.toprak_pH > 7.0:
            print("Toprak pH değeri yüksek. Toprağa kireç veya kül ekleyebilirsiniz.")
        elif self.toprak_pH < 6.0:
            print("Toprak pH değeri düşük. Sulama suyuna asitlik veren malzemeler veya organik malzemeler ekleyebilirsiniz.")
    
    def toprak_nem_kontrol(self):
        if self.toprak_nem >= 60 and self.toprak_nem <= 80:
            print("Toprak nem değeri uygun.")
        elif self.toprak_nem < 60:
            print("Toprak nem değeri düşük. Daha sık sulama yapabilirsiniz.")
        elif self.toprak_nem > 80:
            print("Toprak nem değeri yüksek. Daha az sulama yapabilirsiniz.")
    
    def toprak_sicaklik_kontrol(self):
        if self.toprak_sicaklik >= 10 and self.toprak_sicaklik <= 25:
            print("Toprak sıcaklık değeri uygun.")
        elif self.toprak_sicaklik < 10:
            print("Toprak sıcaklık değeri düşük. Isıtıcı kullanabilirsiniz.")
        elif self.toprak_sicaklik > 25:
            print("Toprak sıcaklık değeri yüksek. Gölgeleme yapabilirsiniz.")
    
    def hava_sicaklik_kontrol(self):
        if self.hava_sicaklik >= 15 and self.hava_sicaklik <= 25:
            print("Hava sıcaklık değeri uygun.")
        elif self.hava_sicaklik < 15:
            print("Hava sıcaklık değeri düşük. Seranın havalandırmasını açabilirsiniz.")
        elif self.hava_sicaklik > 25:
            print("Hava sıcaklık değeri yüksek. Seranın havalandırmasını artırabilirsiniz.")
        
    def sogan_yetistirme_kontrol(self):
        self.toprak_pH_kontrol()
        self.toprak_nem_kontrol()
        self.toprak_sicaklik_kontrol()
        self.hava_sicaklik_kontrol()

tarim_urunu = TarimUrunu()
tarim_urunu.urun_sec()
tarim_urunu.toprak_verilerini_gir()
tarim_urunu.hava_verilerini_gir()

kontrol = SoganYetistirme(tarim_urunu.ph,tarim_urunu.nem,tarim_urunu.toprak_sicaklik,tarim_urunu.hava_sicaklik)   
kontrol.sogan_yetistirme_kontrol()


class FasulyeYetistirme:
    def __init__(self, toprak_pH, toprak_nem, toprak_sicaklik, hava_sicaklik):
        self.toprak_pH = toprak_pH
        self.toprak_nem = toprak_nem
        self.toprak_sicaklik = toprak_sicaklik
        self.hava_sicaklik = hava_sicaklik
        
    def toprak_pH_kontrol(self):
        if self.toprak_pH >= 6.0 and self.toprak_pH <= 7.0:
            print("Toprak pH değeri uygun.")
        elif self.toprak_pH > 7.0:
            print("Toprak pH değeri yüksek. Toprağa kireç veya kül ekleyebilirsiniz.")
        elif self.toprak_pH < 6.0:
            print("Toprak pH değeri düşük. Sulama suyuna asitlik veren malzemeler veya organik malzemeler ekleyebilirsiniz.")
    
    def toprak_nem_kontrol(self):
        if self.toprak_nem >= 50 and self.toprak_nem <= 70:
            print("Toprak nem değeri uygun.")
        elif self.toprak_nem < 50:
            print("Toprak nem değeri düşük. Daha sık sulama yapabilirsiniz.")
        elif self.toprak_nem > 70:
            print("Toprak nem değeri yüksek. Daha az sulama yapabilirsiniz.")
    
    def toprak_sicaklik_kontrol(self):
        if self.toprak_sicaklik >= 15 and self.toprak_sicaklik <= 30:
            print("Toprak sıcaklık değeri uygun.")
        elif self.toprak_sicaklik < 15:
            print("Toprak sıcaklık değeri düşük. Isıtıcı kullanabilirsiniz.")
        elif self.toprak_sicaklik > 30:
            print("Toprak sıcaklık değeri yüksek. Gölgeleme yapabilirsiniz.")
    
    def hava_sicaklik_kontrol(self):
        if self.hava_sicaklik >= 20 and self.hava_sicaklik <= 30:
            print("Hava sıcaklık değeri uygun.")
        elif self.hava_sicaklik < 20:
            print("Hava sıcaklık değeri düşük. Seranın havalandırmasını açabilirsiniz.")
        elif self.hava_sicaklik > 30:
            print("Hava sıcaklık değeri yüksek. Seranın havalandırmasını artırabilirsiniz.")
        
    def fasulye_yetistirme_kontrol(self):
        self.toprak_pH_kontrol()
        self.toprak_nem_kontrol()
        self.toprak_sicaklik_kontrol()
        self.hava_sicaklik_kontrol()

tarim_urunu = TarimUrunu()
tarim_urunu.urun_sec()
tarim_urunu.toprak_verilerini_gir()
tarim_urunu.hava_verilerini_gir()

kontrol = FasulyeYetistirme(tarim_urunu.ph,tarim_urunu.nem,tarim_urunu.toprak_sicaklik,tarim_urunu.hava_sicaklik)   
kontrol.fasulye_yetistirme_kontrol()


class HavucYetistirme:
    def __init__(self, toprak_pH, toprak_nem, toprak_sicaklik, hava_sicaklik):
        self.toprak_pH = toprak_pH
        self.toprak_nem = toprak_nem
        self.toprak_sicaklik = toprak_sicaklik
        self.hava_sicaklik = hava_sicaklik
        
    def toprak_pH_kontrol(self):
        if self.toprak_pH >= 6.0 and self.toprak_pH <= 7.0:
            print("Toprak pH değeri uygun.")
        elif self.toprak_pH > 7.0:
            print("Toprak pH değeri yüksek. Toprağa kireç veya kül ekleyebilirsiniz.")
        elif self.toprak_pH < 6.0:
            print("Toprak pH değeri düşük. Sulama suyuna asitlik veren malzemeler veya organik malzemeler ekleyebilirsiniz.")
    
    def toprak_nem_kontrol(self):
        if self.toprak_nem >= 70 and self.toprak_nem <= 80:
            print("Toprak nem değeri uygun.")
        elif self.toprak_nem < 70:
            print("Toprak nem değeri düşük. Daha sık sulama yapabilirsiniz.")
        elif self.toprak_nem > 80:
            print("Toprak nem değeri yüksek. Daha az sulama yapabilirsiniz.")
    
    def toprak_sicaklik_kontrol(self):
        if self.toprak_sicaklik >= 16 and self.toprak_sicaklik <= 25:
            print("Toprak sıcaklık değeri uygun.")
        elif self.toprak_sicaklik < 16:
            print("Toprak sıcaklık değeri düşük. Isıtıcı kullanabilirsiniz.")
        elif self.toprak_sicaklik > 25:
            print("Toprak sıcaklık değeri yüksek. Gölgeleme yapabilirsiniz.")
    
    def hava_sicaklik_kontrol(self):
        if self.hava_sicaklik >= 20 and self.hava_sicaklik <= 30:
            print("Hava sıcaklık değeri uygun.")
        elif self.hava_sicaklik < 20:
            print("Hava sıcaklık değeri düşük. Seranın havalandırmasını açabilirsiniz.")
        elif self.hava_sicaklik > 30:
            print("Hava sıcaklık değeri yüksek. Seranın havalandırmasını artırabilirsiniz.")
        
    def havuc_yetistirme_kontrol(self):
        self.toprak_pH_kontrol()
        self.toprak_nem_kontrol()
        self.toprak_sicaklik_kontrol()
        self.hava_sicaklik_kontrol()

tarim_urunu = TarimUrunu()
tarim_urunu.urun_sec()
tarim_urunu.toprak_verilerini_gir()
tarim_urunu.hava_verilerini_gir()

kontrol = HavucYetistirme(tarim_urunu.ph,tarim_urunu.nem,tarim_urunu.toprak_sicaklik,tarim_urunu.hava_sicaklik)   
kontrol.havuc_yetistirme_kontrol()


class IspanakYetistirme:
    def __init__(self, toprak_pH, toprak_nem, toprak_sicaklik, hava_sicaklik):
        self.toprak_pH = toprak_pH
        self.toprak_nem = toprak_nem
        self.toprak_sicaklik = toprak_sicaklik
        self.hava_sicaklik = hava_sicaklik
        
    def toprak_pH_kontrol(self):
        if self.toprak_pH >= 6.0 and self.toprak_pH <= 7.5:
            print("Toprak pH değeri uygun.")
        elif self.toprak_pH > 7.5:
            print("Toprak pH değeri yüksek. Toprağa turba veya çam toprağı ekleyebilirsiniz.")
        elif self.toprak_pH < 6.0:
            print("Toprak pH değeri düşük. Toprağa kireç veya kül ekleyebilirsiniz.")
    
    def toprak_nem_kontrol(self):
        if self.toprak_nem >= 50 and self.toprak_nem <= 70:
            print("Toprak nem değeri uygun.")
        elif self.toprak_nem < 50:
            print("Toprak nem değeri düşük. Daha sık sulama yapabilirsiniz.")
        elif self.toprak_nem > 70:
            print("Toprak nem değeri yüksek. Daha az sulama yapabilirsiniz.")
    
    def toprak_sicaklik_kontrol(self):
        if self.toprak_sicaklik >= 10 and self.toprak_sicaklik <= 25:
            print("Toprak sıcaklık değeri uygun.")
        elif self.toprak_sicaklik < 10:
            print("Toprak sıcaklık değeri düşük. Isıtıcı kullanabilirsiniz.")
        elif self.toprak_sicaklik > 25:
            print("Toprak sıcaklık değeri yüksek. Gölgeleme yapabilirsiniz.")
    
    def hava_sicaklik_kontrol(self):
        if self.hava_sicaklik >= 10 and self.hava_sicaklik <= 25:
            print("Hava sıcaklık değeri uygun.")
        elif self.hava_sicaklik < 10:
            print("Hava sıcaklık değeri düşük. Seranın havalandırmasını açabilirsiniz.")
        elif self.hava_sicaklik > 25:
            print("Hava sıcaklık değeri yüksek. Seranın havalandırmasını artırabilirsiniz.")
        
    def ispanak_yetistirme_kontrol(self):
        self.toprak_pH_kontrol()
        self.toprak_nem_kontrol()
        self.toprak_sicaklik_kontrol()
        self.hava_sicaklik_kontrol()

tarim_urunu = TarimUrunu()
tarim_urunu.urun_sec()
tarim_urunu.toprak_verilerini_gir()
tarim_urunu.hava_verilerini_gir()

kontrol = IspanakYetistirme(tarim_urunu.ph,tarim_urunu.nem,tarim_urunu.toprak_sicaklik,tarim_urunu.hava_sicaklik)   
kontrol.ispanak_yetistirme_kontrol()

def main():
    tarim_urunu = TarimUrunu()
    urun_no = tarim_urunu.urun_sec()

    if urun_no == 1:
        karpuz_yetistirme = KarpuzYetistirme()
        karpuz_yetistirme.bilgi_gir(tarim_urunu.ph, tarim_urunu.nem, tarim_urunu.toprak_sicaklik, tarim_urunu.hava_sicaklik)
        karpuz_yetistirme.yetistir()
    elif urun_no == 2:
        cilek_yetistirme = CilekYetistirme()
        cilek_yetistirme.bilgi_gir(tarim_urunu.ph, tarim_urunu.nem, tarim_urunu.toprak_sicaklik, tarim_urunu.hava_sicaklik)
        cilek_yetistirme.yetistir()
    elif urun_no == 3:
        kavun_yetistirme = KavunYetistirme()
        kavun_yetistirme.bilgi_gir(tarim_urunu.ph, tarim_urunu.nem, tarim_urunu.toprak_sicaklik, tarim_urunu.hava_sicaklik)
        kavun_yetistirme.yetistir()
    elif urun_no == 4:
        kayisi_yetistirme = KayisiYetistirme()
        kayisi_yetistirme.bilgi_gir(tarim_urunu.ph, tarim_urunu.nem, tarim_urunu.toprak_sicaklik, tarim_urunu.hava_sicaklik)
        kayisi_yetistirme.yetistir()
    elif urun_no == 5:
        kivi_yetistirme = KiviYetistirme()
        kivi_yetistirme.bilgi_gir(tarim_urunu.ph, tarim_urunu.nem, tarim_urunu.toprak_sicaklik, tarim_urunu.hava_sicaklik)
        kivi_yetistirme.yetistir()
    elif urun_no == 6:
        patates_yetistirme = PatatesYetistirme()
        patates_yetistirme.bilgi_gir(tarim_urunu.ph, tarim_urunu.nem, tarim_urunu.toprak_sicaklik, tarim_urunu.hava_sicaklik)
        patates_yetistirme.yetistir()
    elif urun_no == 7:
        sogan_yetistirme = SoganYetistirme()
        sogan_yetistirme.bilgi_gir(tarim_urunu.ph, tarim_urunu.nem, tarim_urunu.toprak_sicaklik, tarim_urunu.hava_sicaklik)
        sogan_yetistirme.yetistir()
    elif urun_no == 8:
        fasulye_yetistirme = FasulyeYetistirme()
        fasulye_yetistirme.bilgi_gir(tarim_urunu.ph, tarim_urunu.nem, tarim_urunu.toprak_sicaklik, tarim_urunu.hava_sicaklik)
        fasulye_yetistirme.yetistir()     
```
